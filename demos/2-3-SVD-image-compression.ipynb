<div class="mathbook-content"><h1 class="heading"><span class="title">2-3: SVD Image Compression</span></h1><div class="author"><div class="author-name">Robert A. Beezer</div><div class="author-info" /></div></div>

<div class="mathbook-content"><p id="p-1">We use PyLab (included in Sage) to read in an image as an array of grayscale intensities, and save it as an array.  Then we compute the SVD and sum as many rank one terms as determined by the value set by the slider.</p></div>

```python
import pylab
# mean() averages on third axis, the three bytes of color
A_image = pylab.mean(pylab.imread('images/mystery.png'), 2)
@interact
def svd_image(i=(1,(1..50)), display_axes=True):
    u,s,v = pylab.linalg.svd(A_image)
    A     = sum(s[j]*pylab.outer(u[0:,j], v[j,0:]) for j in range(i))
    # g = graphics_array([matrix_plot(A),matrix_plot(A_image)])
    show(matrix_plot(A), axes=display_axes, figsize=(6,8))
    html('Compressed using %s singular values'%i)
```

<div class="mathbook-content"><p id="p-2">We conver to one byte per pixel (8 bits) via an average of the three bytes for the red, green, and blue intensities.  So we have 256 different intensities of black/grey/white.  Each rank-one term requires storing/transmitting a vector with 300 entries and a vector with 500 entries.  For a reconstruction/approximation using the 20 largest singular values, then we require $20\cdot(300+500+1)\approx 15K$ bytes for the image.  This compares with $300\cdot 500\approx 147K$ bytes for the original $500\times 300$ array.</p></div>

<div class="mathbook-content"><p id="p-3">For comparison, maximum lossless compression on this <em class="emphasis">color</em> <abbr class="initialism">PNG</abbr> is 255K in size, a no-compression lossless TIFF format is 440K in size.  Color is probably represented as 3 bytes per pixel with one byte for each of red, green, and blue intensities (256 levels each). A lossy compression <abbr class="initialism">JPEG</abbr> comes in at 71K (equal to about 80 singular values for our black-and-white example).</p></div>
