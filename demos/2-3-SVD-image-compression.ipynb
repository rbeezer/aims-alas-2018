<div class="mathbook-content"><h1 class="heading"><span class="title">2-3: SVD Image Compression</span></h1><div class="author"><div class="author-name">Robert A. Beezer</div><div class="author-info" /></div></div>

<div class="mathbook-content"><p id="p-1">We use PyLab (included in Sage) to read in an image as an array of grayscale intensities, and save it as an array.  Then we compute the SVD and sum as many rank one terms as determined by the value set by the slider.</p></div>

```python
import pylab
A_image = pylab.mean(pylab.imread('images/mystery.png'), 2)
@interact
def svd_image(i=(1,(1..50)), display_axes=True):
    u,s,v = pylab.linalg.svd(A_image)
    A     = sum(s[j]*pylab.outer(u[0:,j], v[j,0:]) for j in range(i))
    # g = graphics_array([matrix_plot(A),matrix_plot(A_image)])
    show(matrix_plot(A), axes=display_axes, figsize=(6,8))
    html('Compressed using %s singular values'%i)
```

<div class="mathbook-content"><p id="p-2">If we use 16 bits (2 bytes) for each pixel, we can have about 65,000 different intensities.  Each rank-one term requires storing/transmitting two vectors with about 300 or 500 entries each.  For 20 terms, then we require $20\cdot(300+500+1)\cdot 2\approx 31K$ bytes for the image.  This compares with perhaps 2 bytes per pixel, over all pixels, for $8\cdot 300\cdot 500\approx 292K$ bytes.  (A PNG color image file for these images run about 68K).</p></div>

<div class="mathbook-content"><p id="p-3">It is worth having a peek at the singular values.</p></div>
