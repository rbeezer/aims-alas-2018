<div class="mathbook-content"><p id="p-2">Set the data points, $(x_i, y_i)\text{,}$ $1\leq i\leq 20\text{.}$</p></div>

```python
data = [(0.54, 5.13), (0.78, 6.97), (1.29, 11.60), (1.78, 10.38), (2.21, 14.15), (2.45, 12.57), (3.33, 17.74), (3.62, 16.12), (3.9, 14.36), (3.92, 17.38), (5.5, 20.32), (6.06, 23.76), (6.42, 24.54), (7.6, 29.61), (7.71, 26.46), (7.92, 26.22), (8.43, 30.88), (8.86, 31.80), (9.22, 33.33), (9.22, 33.59)]
```

<div class="mathbook-content"><p id="p-3">Explore the data to look for a trend.</p></div>

```python
data_plot = list_plot(data)
data_plot
```

<div class="mathbook-content"><p id="p-4">Employ a linear <em class="emphasis">model</em>, $y = ax + b$ to a system of $20$ equations in the unknown values of the model parameters $a$ and $b\text{.}$</p></div>

```python
A = matrix(RDF, [[p[0], 1] for p in data])
b = vector(RDF, [p[1] for p in data])
print A.str()
print b.column().str()
```

<div class="mathbook-content"><p id="p-5">Construct the normal equations $\adjoint{A}Ax=\adjoint{A}b$ and solve.</p></div>

```python
s = (A.conjugate_transpose()*A).solve_right(A.conjugate_transpose()*b)
s
```

<div class="mathbook-content"><p id="p-6">Compute the residuals.</p></div>

```python
A*s - b
```

<div class="mathbook-content"><p id="p-7">Draw the model along with the data.</p></div>
