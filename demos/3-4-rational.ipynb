<div class="mathbook-content"><p id="p-2">We have seen the two eigenvalues before.</p></div>

```python
A.eigenvalues()
```

<div class="mathbook-content"><p id="p-3">Sage does characteristic polynomials, and factored characteristic polynomials.</p></div>

```python
p = A.charpoly()
p
```

```python
A.fcp()
```

<div class="mathbook-content"><p id="p-4">Cayley-Hamilton says:</p></div>

```python
p(A)
```

<div class="mathbook-content"><p id="p-5">Minimal polynomials, too.</p></div>

```python
m = A.minimal_polynomial()
```

<div class="mathbook-content"><p id="p-6">By definition:</p></div>

```python
m(A)
```

<div class="mathbook-content"><p id="p-7">Minimal is a divisor of characteristic:</p></div>

```python
m.divides(p)
```

<div class="mathbook-content"><p id="p-8">So if we factor $m(x)$ we should still have linear factors expressing the eigenvalues.</p></div>

```python
m.factor()
```

<div class="mathbook-content"><p id="p-9">Where have we seen those exponents before? (Hint: index of nilpotent matrices.)</p></div>

<div class="mathbook-content"><p id="p-10">So we know a lot about this matrix.  Finally—the rational canonical form.</p></div>

```python
A.rational_form()
```
